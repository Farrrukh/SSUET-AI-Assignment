### OPTIONAL ARGUMENT

```python
def name(fname,lname,mname=""):
    full_name = fname+" "+mname+" "+lname
    print(full_name)
    print(fname)
    print(mname)
    print(lname)
name("Ramsha","Azeemi",mname='M')
```

### UNKNOWN NO. OF ARGUMENTS


#### *  -> tuple (takes positional arg.)
#### ** -> Dictionary (takes keyword arg.)

```python
def customized_gift(msg,counter,size,*my_object):      #Parameters
        print("Recalling your order")
        print("The object you have ordered are :\n")
        for item in my_object:
            print("#",item)
        print("You custom message :",msg)
        print("Size of the object",size)
        print("No. of items ",counter)
    
customized_gift("Pakistan Zindabad" , 1 , "Medium" , "T shirt" , "Mug" , "Frame" )    # Arguments   
```

```python
def customized_gift(msg,counter,*size,**my_object):      #Parameters
        print("Recalling your order")
        print("The object you have ordered are :\n")
        for key,value in my_object.items():
            print("key : ",key)
            print("Value: ",value)
        print("You custom message :",msg)
        print("Size of the object",size)
        print("No. of items ",counter)
        print("\n\n\n\n")
        print(my_object)
    
customized_gift("Pakistan Zindabad" , 1 ,"Medium","large","XL", shirt = "T shirt" , item2 ="Mug" , item3 ="Frame" , item4='xyz')    # Arguments   
```
